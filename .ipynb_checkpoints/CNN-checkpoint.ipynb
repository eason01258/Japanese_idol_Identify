```python
data_path = '.\\training_set'

x_data_list = []
y_data_list = []
for roots, dirs, files in os.walk(data_path):  #遞迴指定路徑下所有的子目錄與檔案
#     print("路徑：", roots)
#     print("目錄：", dirs)
#     print("檔案：", files)
#     print('---------------------\n')
    if dirs == []:  #遞迴至目錄內(該目錄內已沒有其他子目錄)
        #print('null',dirs)  
        for each in files:
            #print('every each:',each)
            if each.find('check') == -1: #如果檔案名稱不含有字串'check'
                x_data_list.append(os.path.join(roots, each))
                y_data_list.append(roots.split("\\")[-1]) 
```

```python
print(x_data_list[0:5])
#print(y_data_list)
```

```python
img_size = 224  
img = cv2.imread(x_data_list[7],1) # (1 BGR三個 channels 的彩色圖片，而忽略透明度的 channel。), ( 0 灰階 ) ,( -1 讀取圖片中所有的 channels，包含透明度的 channel。)
print(img.shape)  # 圖片原始尺吋 316 * 263
plt.imshow(img) # BGR
plt.axis('off')
plt.show()
img = img[:,:,::-1] #將 BGR 轉成 RGB
img = cv2.resize(img, (img_size, img_size)) #將圖片大小調整成 224 * 224 (img_size = 224)
print(img.shape)
plt.imshow(img) # 顯示圖片
plt.axis('off') #不顯示軸
plt.show()
```

```python
from tqdm import tqdm # 載入進度條
def load_img(data_list):
    data_img = []
    for each in tqdm(data_list):
        img = cv2.imread(each, 1) # 1 讀取RGB
        img = img[:,:,::-1] #轉向 BGR --> RGB
        img = cv2.resize(img, (img_size, img_size)) 
        #data_img.append(img[..., np.newaxis]) #增加維度 
        data_img.append(img)
    return np.array(data_img).astype('float32')/255 # 讓圖片介於 0~1 之間
```

```python
x_data = load_img(x_data_list) # 載入圖片集 
```

```python
for i in range(5):
    plt.figure(figsize=(16, 8))
    plt.subplot(2, 4, i+1)
    print(x_data[i].shape)
    plt.imshow(x_data[i],cmap ='gray' )
    plt.axis('off')
plt.show()
```

```python
y_data_list = pd.DataFrame(y_data_list, columns=['label']) #將 y_data_list 轉成 DataFrame 格式 column 名為 label
print(y_data_list['label'].value_counts()) # 顯示各個類別資料數
uniques = y_data_list['label'].value_counts().index  #顯示class  或 y_data_list['label'].unique()
#uniques = y_data_list['label'].unique() #同上
print(uniques)
counts = y_data_list['label'].value_counts().values #顯示資料數
print(counts)
```

```python
# 讀取 class map
class_path = "./classmap.csv"
class_map = pd.read_csv(class_path,header = None, index_col = 0)
class_map
```

```python
# class map 轉成字典
class_map = class_map.to_dict()
class_map1 = class_map[1]
class_map1
```

```python
# map label 
y_data = y_data_list['label'].map(class_map1).values.copy() #values.copy 只複製值
y_data
```

```python
# 切分測試集驗證集

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data,
                                                    test_size=0.2,
                                                    random_state=42,
                                                    stratify=y_data)
```

```python
y_test[:5]
```

```python
# 将類別向量做ont-not encoding
from keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
num_classes = len(uniques)
y_test[:5]
```

```python
x_train.shape # 四個維度 1: 比數 , 2、3 : size 224 *224 , 4 : RGB 三原色 為3
```

```python
x_train.shape[1:] #(圖片長 , 圖瑱寬, chanel (RGB) )
```

```python
inputs = Input(shape=x_train.shape[1:]) 
    
x = layers.Conv2D(32, 3, padding = 'same')(inputs)
x = layers.Conv2D(32, 3, padding = 'same')(x)
x = layers.ReLU()(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D((2,2))(x)
    
    
x = layers.Conv2D(64, 3, padding = 'same')(x)
x = layers.Conv2D(64, 3, padding = 'same')(x)
x = layers.ReLU()(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D((2,2))(x)
    
x = layers.Conv2D(128, 3, padding = 'same')(x)
x = layers.Conv2D(128, 3, padding = 'same')(x)
x = layers.ReLU()(x)
x = layers.BatchNormalization()(x)
    
    
x = layers.AveragePooling2D(pool_size=(3,3))(x)
#     x = layers.GlobalAveragePooling2D()(x)
x = layers.Flatten()(x)
x = layers.Dropout(0.4)(x)
x = layers.Dense(7, activation='softmax')(x)

outputs = Dense(num_classes, activation='softmax')(x)


model_2 = Model(inputs=inputs, outputs=outputs)
```

```python
batch_size = 25 # 每看25筆資料更新一次權重
epochs = 70 #迭帶50次
```

```python
opt_adam = keras.optimizers.Adam(lr=0.0001,decay=1e-6) #learn rate 0.0001 # 每次更新後learn衰減10^-6
model_2.compile(loss='categorical_crossentropy',
        optimizer = opt_adam,
        metrics = ['accuracy'])
```

```python
model_dir = './Model/model_2-logs'
if not os.path.exists(model_dir):
    os.makedirs(model_dir)

logfiles = model_dir + '/{}-{}'.format('basic_model',
                                       model_2.__class__.__name__)
model_cbk = keras.callbacks.TensorBoard(log_dir=logfiles,
                                        histogram_freq=1)

modelfiles = model_dir + '/{}-best-model.h5'.format('basic_model')
model_mckp = keras.callbacks.ModelCheckpoint(modelfiles,
                                             monitor='val_accuracy',
                                             save_best_only=True)

earlystop = keras.callbacks.EarlyStopping(monitor='val_loss',
                                          patience=5,
                                          verbose=1)


callbacks_list = [model_cbk, model_mckp, earlystop]
```

```python
num_step = x_train.shape[0] // batch_size
```

```python
datagen = ImageDataGenerator(
                             width_shift_range=0.1,
                             height_shift_range=0.1,
                             horizontal_flip=True,
                             vertical_flip=False,
                             fill_mode='nearest',
                             cval=0,
                             )


history_2 = model_2.fit_generator(datagen.flow(x_train, y_train,
                                               batch_size=batch_size),
                                  steps_per_epoch=num_step,
                                  epochs=epochs,
                                  validation_data=(x_test, y_test),
#                                   callbacks=callbacks_list
                                 )
```

```python
y_pred = model_2.predict(x_test)
y_pred = y_pred.argmax(-1)
```

```python
class_names = np.array([each for each in class_map.keys()])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print(accuracy_score(y_test.argmax(-1), y_pred))
print(confusion_matrix(y_test.argmax(-1), y_pred))
```

```python
import seaborn as sns

matrix = confusion_matrix(y_test.argmax(-1), y_pred)

plt.figure(figsize=(10, 10))
sns.heatmap(matrix, annot=True, cbar=False, cmap="Blues",
            xticklabels=class_names,
            yticklabels=class_names)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')
plt.show()
```
